```python
def joukowski_transform(zeta,c):
    """Returns the Joukowski transform of a complex number on plane zeta
    
    Arguments
    ---------
    zeta = xi + j*eta
    
    Returns
    -------
    x: x-coordinate after transform
    y: y-coordinate after transform
    """
    
    z = zeta + c**2/zeta
    x = numpy.real(z)
    y = numpy.imag(z)
    
    return x, y
    
```

```python
#Section 2.1
%matplotlib inline
c=1
R=1.5
theta = numpy.linspace(0,2*math.pi,150)
z = R*numpy.cos(theta) + 1j*R*numpy.sin(theta)
x_1 = numpy.real(z)
y_1 = numpy.imag(z)
x_trans, y_trans = joukowski_transform(z,c)
size = 6
pyplot.figure(figsize=(size,size))
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim(-3.0,3.0)
pyplot.ylim(-3.0,3.0)
pyplot.title('Original + Transform')
pyplot.plot(x_1, y_1, color='#CD2305', linewidth=2, linestyle='--')
pyplot.plot(x_trans,y_trans, color='b', linewidth=2, linestyle='-')
```
