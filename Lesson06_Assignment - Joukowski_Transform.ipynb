```python
def joukowski_transform(zeta,c):
    """Returns the Joukowski transform of a complex number on plane zeta
    
    Arguments
    ---------
    zeta = xi + j*eta
    
    Returns
    -------
    x: x-coordinate after transform
    y: y-coordinate after transform
    """
    
    z = zeta + c**2/zeta
    x = numpy.real(z)
    y = numpy.imag(z)
    
    return x, y
    
```

```python
def get_velocity_doublet(strength, xd, yd, X, Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    u = - strength/(2*math.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2*math.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
    return u, v
```

```python
U_inf = 1.0
kappa = R**2*2*math.pi*U_inf

uz, vz = get_velocity_doublet(kappa,xc,yc,x,y)
uz = uz + U_inf

# Complex Analysis
u = uz - 1j*vz
dEdz = 1 - (c/z)**2

WE = u/dEdz

uE = numpy.real(WE)
vE = -numpy.imag(WE)
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
    """Returns the velocity field generated by a vortex.
    
    Arguments
    ---------
    strength -- strength of the vortex.
    xv, yv -- coordinates of the vortex.
    X, Y -- mesh grid.
    """
    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u, v
```

```python
# Calculate vortex strength to move stagnation points to theta = 0
gamma = -math.sin(-AoA)*4*math.pi*R
print 'gamma =', gamma

uprime_doublet, vprime_doublet = get_velocity_doublet(kappa, xcprime, ycprime, xprime, yprime)
uprime_vortex, vprime_vortex = get_velocity_vortex(gamma, xcprime, ycprime, xprime, yprime)

uprime = uprime_doublet + uprime_vortex + U_inf
vprime = vprime_doublet + vprime_vortex


# Complex Analysis
Uprime = uprime - 1j*vprime
dEdz = 1 - (c/z)**2

WE = Uprime/dEdz

#Transform back to original coordinate system
U = Uprime*numpy.exp(-1j*AoA)
EU = WE*numpy.exp(-1j*AoA)

uz = numpy.real(U)
vz = -numpy.imag(U)

uE = numpy.real(EU)
vE = -numpy.imag(EU)
```
