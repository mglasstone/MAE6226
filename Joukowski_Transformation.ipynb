```python
import numpy 
import math
import cmath
from matplotlib import pyplot
```

```python
theta = numpy.linspace(0,2*math.pi, 100)
R_1 = 1.0
# use complex numbers to define coords
zeta = R_1*numpy.cos(theta) +1j* R_1*numpy.sin(theta)
# for plotting
x_1 = numpy.real(zeta)
y_1 = numpy.imag(zeta)
```

```python
def joukowski_transform(zeta):
    """Returns the Joukowski transform of a complex number on plane zeta
    
    Arguments
    ---------
    zeta = xi + j*eta
    
    Returns
    -------
    x: x-coordinate after transform
    y: y-coordinate after transform
    """
    
    z = zeta + 1/zeta
    x = numpy.real(z)
    y = numpy.imag(z)
    
    return x, y
    
```

```python
#perform transform on R=1 circle
x_1, y_1 = joukowski_transform(zeta)

%matplotlib inline
size = 6
pyplot.figure(figsize=(size,size))
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim(-2.0,2.0)
pyplot.ylim(-2.0,2.0)
pyplot.title('Transformed Circle')
pyplot.plot(x_1, y_1, color='#CD2305', linewidth=2, linestyle='-')
```

```python
# circle of R_3 = 1.5, centered at -0.5,0
R_3 = 1.5
zeta_3 = -0.5+R_3*numpy.cos(theta) + 1j*R_3*numpy.sin(theta)
x_3, y_3 = joukowski_transform(zeta_3)


size = 6
pyplot.figure(figsize=(6,4))
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim(-3.0,3.0)
pyplot.ylim(-2.0,2.0)
pyplot.title('Transformed Circle')
pyplot.plot(x_3, y_3, color='#CD2305', linewidth=2, linestyle='-')
```

```python
# implement a doublet of radius 1.5 at -0.5, 0.0

# generate mesh
N = 50                                # Number of points in each direction
x_start, x_end = -2.0, 2.0            # x-direction boundaries
y_start, y_end = -1.0, 1.0            # y-direction boundaries
x = numpy.linspace(x_start, x_end, N)    # computes a 1D-array for x
y = numpy.linspace(y_start, y_end, N)    # computes a 1D-array for y
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid
```

```python
U_inf = 1.0 # free stream flow speed

kappa = 2*math.pi*U_inf*R_3**2
x_doublet, y_doublet = -0.5, 0.0
```

```python
def get_stream_function_doublet(strength, xd, yd, X, Y):
    """Returns the stream-function generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    psi = - strength/(2*math.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    
    return psi

def get_velocity_doublet(strength, xd, yd, X, Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    u = - strength/(2*math.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2*math.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
    return u, v
```

```python
psi = get_stream_function_doublet(kappa, x_doublet, y_doublet, X, Y)
u, v = get_velocity_doublet(kappa, x_doublet, y_doublet, X, Y)

coords = X + 1j*Y
velocity = u + 1j*v
X_trans, Y_trans = joukowski_transform(coords)
u_trans, v_trans = joukowski_transform(coords)
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-2.5, 2.5)
pyplot.ylim(-2.5, 2.5)
pyplot.scatter(X_trans, Y_trans, s=10, color='#CD2305', marker='o', linewidth=0)
```
